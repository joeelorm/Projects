```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

```python
iris = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris.head()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
iris_pca = PCA(n_components = 4)
components = iris_pca.fit(iris).components_
components = pd.DataFrame(components).transpose()
components.columns = ['Comp1', 'Comp2', 'Comp3', 'Comp4']
components.index = iris.columns
print(components)
```

```python
var_ratio = iris_pca.explained_variance_ratio_
var_ratio = pd.DataFrame(var_ratio).transpose()
var_ratio.columns = ['Comp1', 'Comp2', 'Comp3', 'Comp4']
var_ratio.index = ['Proportion of Variation']
print(var_ratio)
```

```python
iris_pcomp = iris_pca.fit_transform(iris)
iris_pcomp = pd.DataFrame(iris_pcomp)
iris_pcomp = iris_pcomp.iloc[:, 0:3]
iris_pcomp.columns = ['Comp1', 'Comp2', 'Comp3']
print(iris_pcomp.head())
```

```python
sns.pairplot(iris_pcomp)
plt.show()
```
